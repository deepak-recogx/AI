```python
import pandas as pd
import numpy as np

reviews_datasets = pd.read_csv(r'D:\RecogX_Backup\Dataset\Reviews.csv')
reviews_datasets = reviews_datasets.head(20000)
```

```python
reviews_datasets.dropna()
```

```python
reviews_datasets.head()
```

```python
#Distribution of rating:
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.countplot(x='Score', data=reviews_datasets)
```

```python
reviews_datasets['Text'][200]
```

```python
#Subjectivity quantifies the amount of personal opinion and factual information 
#contained in the text. The higher subjectivity means that the text 
#contains personal opinion rather than factual information.
```

```python
text_blob_object = TextBlob(reviews_datasets['Text'][200])
print(text_blob_object.sentiment)
```

```python
def find_pol(review):
    return TextBlob(review).sentiment.polarity

reviews_datasets['Sentiment_Polarity'] = reviews_datasets['Text'].apply(find_pol)
reviews_datasets.head()
```

```python
#Distribution of polarity
sns.distplot(reviews_datasets['Sentiment_Polarity'])
```

```python
#Average polarity for each score rating.
sns.barplot(x='Score', y='Sentiment_Polarity', data=reviews_datasets)
```

```python
#Negative reviews
most_negative = reviews_datasets[reviews_datasets.Sentiment_Polarity == -1].Text.head()
print(most_negative)
```
